```python
# Today, we will create a function that flattens a list
output = []
def flatten(liste):
    for i in liste:
        if type(i) == list:
            flatten(i)
        else:
            output.append(i)
    return output
```

```python
# input sample
flatten([[1, 'a', ['cat'], 2], [[[3]], 'dog'], 4, 5]) 
```
